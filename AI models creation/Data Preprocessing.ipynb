```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
import string
import random
```

```python
nltk.download('punkt')
```

```python
df = pd.read_csv('/content/Procurement.csv')
rows, columns = df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
df = df[['English Text', 'Arabic Text']]
```

```python
df
```

```python
def normalize_text(text):
    # Convert the text to lowercase
    text = text.lower()
    # Replace multiple spaces with a single space and strip leading/trailing spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove text inside parentheses
    text = re.sub(r'\(.*?\)', '', text)
    # Tokenize the text into words
    words = word_tokenize(text)
    return ' '.join(words)
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df['English Text'] = df['English Text'].apply(normalize_text)
df['Arabic Text'] = df['Arabic Text'].apply(normalize_text)
```

```python
df
```

# **Sentences df**

```python
# Funtion to extract only sentences
def sentence_only(df):
  # len(x.split()) gives the number of words and we only keeping sentences of minimum 5 words
  sentence_df = df[df['English Text'].apply(lambda x: len(x.split()) >= 5)]
  return sentence_df
```

```python
sentences_df = sentence_only(df)
rows, columns = sentences_df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
def concatenate_multiple_sentences(df, max_sentences):
    # List to hold the concatenated sentences
    concatenated_rows = []
    # Iterate over the range from 2 to max_sentences
    for num_sentences in range(2, max_sentences + 1):
        for i in range(len(df) - num_sentences + 1):
            english_concat = ' '.join(df['English Text'].iloc[i:i + num_sentences])
            arabic_concat = ' '.join(df['Arabic Text'].iloc[i:i + num_sentences])
            concatenated_rows.append({'English Text': english_concat, 'Arabic Text': arabic_concat})
    # Create a new DataFrame from the concatenated rows
    concatenated_df = pd.DataFrame(concatenated_rows)
    return concatenated_df
```

```python
concat_sentences = concatenate_multiple_sentences(sentences_df, 10)
rows, columns = concat_sentences.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
concat_sentences
```

# **Words df**

```python
# Funtion to extract only words
def words_only(df):
  words_df = df[df['English Text'].apply(lambda x: len(x.split()) <= 10)]
  return words_df
```

```python
words_df = words_only(df)
rows, columns = words_df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
concat_words = concatenate_multiple_sentences(words_df, 10)
rows, columns = concat_words.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

# **Combine all df**

```python
combined_df = pd.concat([concat_sentences, concat_words]).reset_index(drop=True)
rows, columns = combined_df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
combined_df = pd.concat([combined_df, df]).reset_index(drop=True)
rows, columns = combined_df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
combined_df = combined_df.sample(frac=1).reset_index(drop=True)
```

```python
rows, columns = combined_df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```
